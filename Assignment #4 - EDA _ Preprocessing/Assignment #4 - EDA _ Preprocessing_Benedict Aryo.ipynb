```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('iris.csv')
data
```

## Soal 1 (Bobot: 30)

Buatlah scatter plot dari data Iris yang menampilkan hubungan antara kolom "sepal length" dan "petal length"

### Jawab :

```python
# Implementasi kode scatter plot di sini
import matplotlib.pyplot as plt
%matplotlib inline
plt.scatter(data['sepal length'],data['petal length'] )
plt.title('Persebaran korelasi "sepal length" dan "petal length"')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.show;
```

Jelaskan makna dari scatter plot tersebut! 

<font color=red>Tulis jawaban anda di bawah!</font>

### Jawab :

```python
import numpy as np
np.corrcoef(data['sepal length'],data['petal length'])
```

__Sepal Length dan Petal Length memiliki korelasi positif <br> dimana semakin tinggi nilai Sepal Length, maka nilai Petal Length semakin tinggi juga__
